## Import e função de scraping

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options

from bs4 import BeautifulSoup
import re

import datetime
import time

import pandas as pd

option = webdriver.ChromeOptions()

#Argumentos para "driblar" o bloqueio de bots
option.add_argument('--disable-blink-features=AutomationControlled')
option.add_argument("window-size=1280,800")
option.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36")
option.add_argument(r"user-data-dir=C:\Users\lgrne\OneDrive\Documents\AULAS PUC BI\Projeto Final\selenium\Default")

s = Service(r'C:\Users\lgrne\Documents\chromedriver.exe')
```

```python
def skyscanner_precos_ber(ida, volta, sleep_time=2):

    precos = []
    
    # Manipulação da url de busca. Podemos editar origem, destino e datas à vontade aqui.
    
    url = "https://www.skyscanner.com.br/transporte/passagens-aereas/rioa/ber/" + str(ida) + "/" +\
           str(volta) + "/?adults=1&adultsv2=1&cabinclass=economy&children=0&childrenv2=&destination"\
           "entityid=27547053&inboundaltsenabled=false&infants=0&originentityid=27541837&outboundaltsenabled=false"\
           "&preferdirects=false&ref=home&rtn=1"
    
    driver.get("https://www.skyscanner.com.br/")
    time.sleep(15) #Há trechos de sleep ao longo do código para ajudar a "driblar" a detecção de bots

    driver.get(url)
    time.sleep(20)
      
    soup = BeautifulSoup(driver.page_source, "html.parser")

    # O site do Skyscanner fica carregando os resultados durante um tempo variável. 
    # Esse trecho garante que só busquemos o valor mais barato quando de fato ele terminar de carregar. 
    # Enquanto a página não for carregada corretamente, um aviso "Página ainda não carregada" será exibido.
    
    while ("resultados" in str(soup.select('div[class*="ResultsSummary"]'))) == False:
        print("Página ainda não carregada")
        time.sleep(10)
        soup = BeautifulSoup(driver.page_source, "html.parser")
     
    
    for EachPart in soup.select('div[class*="mainPrice"]'):
        precos.append(EachPart.get_text())
        
    return precos[1]

```

## Scraping

```python
# Lista onde os valores serão armazenados

lista_2022 = []
```

```python
## Para 2022
dias = [1,3,6,9,12,15,18,21]
meses = [7,8,9,10,11,12]
```

```python
# Loop de coleta para cada dia e mês explicitado nas variáveis "dias" e "meses"

for mes in meses:
    for dia in dias:
        t = (2022,mes,dia,0,0,0,0,0,0)
        secs = time.mktime(t)
        
        data_ida = datetime.datetime.fromtimestamp(secs)
        data_volta = data_ida + datetime.timedelta(days=7)
        
        driver = webdriver.Chrome(service=s, options=option)
        time.sleep(5)
        preco = skyscanner_precos_ber(data_ida.strftime("%Y%m%d"),data_volta.strftime("%Y%m%d"))
        time.sleep(5)
        driver.quit()
        
        consulta = {'Destino': 'Berlim', 'Ida': data_ida, 'Volta': data_volta, 'Preco': preco}
        lista_2022.append(consulta)
```

```python
# Criação do dataframe

df_2022 = pd.DataFrame(lista_2022)
```
